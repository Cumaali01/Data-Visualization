# Functional Method and Object Oriented Methods

## Functional Method (2 rows 1 column)

```python
age = np.arange(25,44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 48876, 53850, 57287,
            65998, 71496, 75370, 83640]
```

```python
plt.figure(figsize=(6,4))
plt.style.use("seaborn-darkgrid")

plt.plot(age,salary)
plt.xlabel("Age", color="b")
plt.ylabel("Salary", color="purple")
plt.title("SALARY BY AGE", color="red");

plt.grid(color="purple")
```

## OOM

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(age, salary, "g+-")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("SALARY BY AGE");
```

```python
plt.subplot(2, 1, 1)
plt.plot(age,salary, "r")

plt.subplot(2, 1, 2)
plt.plot(age, salary_2, "b");
```

## OOM Multi Plot (2 rows 1 column)

```python
fig, ax = plt.subplots(2,1)

ax[0].plot(age, salary, "r")

ax[1].plot(age, salary_2, "g")

plt.tight_layout(); # not to mix two graphs
```

## Functional Method (2 rows 3 columns, total 6 axes)

```python
mpl.rcParams.update(mpl.rcParamsDefault)

plt.subplot(2,3,1)
plt.plot(age,salary, "r")

plt.subplot(2,3,2)
plt.plot(age, salary2, "g")

plt.subplot(2,3,3)
plt.plot(age, salary, "y")

plt.subplot(2,3,4)
plt.plot(age, salary2, "y")

plt.subplot(2,3,5)
plt.plot(age, salary, "bo")

plt.subplot(2,3,6)
plt.plot(age, salary2, "b:")

plt.tight_layout()
plt.show()
```

## OOM Multi Plot (2 rows 3 columns)

```python
mpl.rcParams.update(mpl.rcParamsDefault) ## Grafiklerdeki tüm modifikasyonları default değerlere döndürdük
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3)
ax[0, 0].plot(age, salary, "r")
ax[0, 1].plot(age, salary2, "grey")
ax[0, 2].plot(age, salary, "bo")
ax[1, 0].plot(age, salary2, "pink", linestyle="dashed", linewidth=4)
ax[1, 1].plot(age, salary, "purple")
ax[1, 2].plot(age, salary2, "green")
fig.tight_layout()
plt.show()
```

```python
x = np.arange(0,30,.2)

fig, ax = plt.subplots(figsize= (14,8))

ax.plot(x, x-1, color="green", linewidth=3, linestyle="-")
ax.plot(x, x-2, color="red", lw=3, linestyle="-.")
ax.plot(x, x-3, color="yellow", lw=3, ls=":")
ax.plot(x, x-4, color="purple", lw=3, ls="dashdot")

plt.show()
```

```python
x = np.arange(10)

fig, ax = plt.subplots(2,2, figsize=(12,6))

plt.suptitle("LineStyle &  Markers & MarkerSize & Color", fontsize= 16, y= 0.95, color="r", fontweight="bold", fontstyle="italic") # y= 0.98 is default for title location

ax[0,0].plot(x, "r", marker= "+", markersize = 5)
ax[0,1].plot(x-2, "y", marker= "o", ms = 5)

ax[1,0].plot(x-3, "g", marker= "o", ms = 5, lw=3)
ax[1,1].plot(x-4, "b", marker= "1", ms = 5, lw=5)

plt.show()
```

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)
fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))
ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.show()
```

# Pie Chart

```python
x = np.array([100,100,100,100])
my_labels= ["Apples", "Bananas", "Cherries","Dates"]

plt.figure(figsize=(7,7))
plt.pie(x,labels=my_labels, labeldistance=-0.25, autopct="%.2f");
```

```python
z = np.array([16,21,45,19])
my_labels= ["Apples", "Bananas", "Cherries","Dates"]

myexplode= [0.1, 0, 0, 0]
```

```python
plt.figure(figsize=(7,7))

plt.pie(z, labels=my_labels, explode= myexplode, labeldistance=0.75, autopct="%.2f", startangle=45);
```

```python
# %matplotlib notebook
%matplotlib inline
plt.figure(figsize=(10,8))

color_list = ["gold", "yellowgreen", "lightcoral","lightskyblue"]
explode_list = [0.075, 0.02, 0.1, 0.05]
my_labels = ["Apples", "Bananas", "Cherries","Pear"]

y = np.array([25,35,25,15])

plt.pie(y, labels=my_labels, explode=explode_list, labeldistance=.7, autopct="%.2f", startangle=90, shadow=True, pctdistance=0.6, colors=color_list)

plt.legend(loc="best")
plt.show()
```
