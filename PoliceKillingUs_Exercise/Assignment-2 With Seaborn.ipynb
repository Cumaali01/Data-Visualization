## 1a: Box plot of age in both genders separeted by manner of death.

## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

## 2: Plot 12 most dangerous cities

```python
kill.city.value_counts().head(12) #1st answer
```

```python
kill.city.value_counts()[:12] #2nd answer
```

```python
cty = kill.city.value_counts().nlargest(12) #3rd answer
cty
```

```python
cty.values
```

```python
cty.index
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.barplot(x=cty.index, y=cty.values)
ax.set_xticklabels(cty.index, color="b", fontsize=8, rotation=45)
ax.set_title("Most Dangerous City", color="r");
```

```python
# with seaborn
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
kill_state = kill.state.value_counts().head(20)
kill_state
```

```python
kill_state.values
```

```python
# with seaborn barplot

fig, ax = plt.subplots(figsize=(10,6))

sns.barplot(x=kill_state.index, y=kill_state.values)
ax.set_xticklabels(kill_state.index, color="b", fontsize=8, rotation=45,)
ax.set_title("Most Dangerous States", color="r", fontsize=18, fontweight="bold",fontstyle="italic")
ax.set_xlabel("states", color="purple", fontsize=20)
ax.set_ylabel("kills", color="r", fontsize=18)
plt.show()
```

```python
# with seaborn countplot

# fig, ax = plt.subplots(figsize=(10,6))

sns.countplot(x=kill_state.values, data=kill)
# ax.set_xticklabels(kill_state.index, color="b", fontsize=8, rotation=45,)
# ax.set_title("Most Dangerous States", color="r");
plt.show()
```
